### Linear Regression Theory
    - Paramatric
    - İki değişken arasındaki ilişkiyi göstermek için linear regression doğrusu kullanılır.
    - Linear regresyonda bir değişken diğerini etkiler.
    - Statistical regression method used for predictive analysis.
    - y = ax + b
    - Hataların kareleri toplamını minimize ederek best fit line' ı çizeriz.
    - En küçük kareler yöntemi(least square) kullanılarek hesaplama yapılır.
    - Algoritma olarak Gradient Descent modeli kullanılır.
    

### Linear Regression Assumptions
    - Feature'lar arasında güçlü bir lineer ilişki olması gerekir.
    - Hataların normal(homojen) dağılım sergilemesi gerekir.
### Regression Error Metrics
    1. Mean Absolute Error(MAE): Hataları cezalandırmaz.
    2. Mean Squared Error(MSE): Küçük hataları bile cezalandırır.
    3. Root Mean Squared Error(RMSE): İdeal metric.

##### Error Metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
R2_score = r2_score(y_test, y_pred)
R2_score
# modelimiz datamızı %90 oranında açıklıyor. % 10 luk kayıp verimiz var. 
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
# mean absolute error.hataları cezalandırmaz
```

```python
mse = mean_squared_error (y_test, y_pred)
mse
# mean squared error. küçük hataları bile çok fazla cezalandırır ve yorumlaması zor.
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse
# root mean squared error. en uygun metric.
```

##### Adjusted R^2 Score

```python
def adjusted_r2(y_test, y_pred, X):
    r2 = r2_score(y_test, y_pred)
    n = X.shape[0] # number of observations
    p = X.shape[1] - 1 # number of independent variables
    adjusted_r2 = 1 - (1-r2) * (n-1) / (n-p-1)
    return adjusted_r2

# feature sayısı artınca R^2' de yalancı bir iyileşme oluyor. Bunu engellemek için bu fonksiyon kullanılıyor.
# feature sayısı ile gözlem sayısı doğru orantılı olarak artmalı ki yalancı bir iyileşme olmasın. Normal bir iyileşme olsun.
# Sklearn kütüphanesi içinde böyle bir fonksiyon olmadığı için bu fonksiyonu biz tanımladık.
# Büyük data olsun küçük data olsun adjusted_r2 bakmakta fayda var.
```

```python
adjusted_r2(y_test, y_pred, X)
```

```python
R2_score = r2_score(y_test, y_pred)
R2_score
# modelimiz datamızı %90 oranında açıklıyor. % 10 luk kayıp verimiz var.
```

##### Model Evaluation

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mse)
    R2_score = r2_score(actual, pred)
    print("Model testing performance:")
    print("--------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
```

```python
eval_metric(y_test, y_pred)
```

```python
y_train_pred = model.predict(X_train)
```

```python
eval_metric(y_train, y_train_pred)
```

##### Plotting Residuals

```python
residuals = y_test - y_pred
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x = y_test, y = residuals)
plt.axhline(y = 0, color = "r", linestyle="--")
plt.ylabel("residuals")
plt.show()
# Hataların normal bir dağılım sergilediğiini görebiliyoruz.
# Böylece lineer regresyon için gerekli olan ikinci varsayımı da sağladığını görmüş olduk.
```

```python
sns.kdeplot(residuals)
# kde plot'a bakarak da hataların normal dağılıma yakın olduğunu görebiliyoruz.
```

```python
stats.probplot(residuals, dist="norm", plot=plt);
# Bu grafikte de hataların çizgi etrafında yılan gibi dolandığını görebiliyoruz.
# Bu da hataların normal dağılım sergilediğini gösteriyor.
```

```python
from scipy.stats import skew
skew(residuals)
# -0.5 ile 0.5 arasındaki değerler datanın normal dağılıma yaklaştığını gösterir.
```

```python
from yellowbrick.regressor import ResidualsPlot

# Instantiate the linear model and visalizer
model = LinearRegression()
visualizer = ResidualsPlot(model)

visualizer.fit(X_train, y_train) # Fit the training data to a visualizer.
visualizer.score(X_test, y_test) # Evaluate the model on the test data
visualizer.show();   # Finalize and render the figure

# yellowbrick kütüphanesi kullanarak da datamızın normal dağılıma yakın olduğunu gördük.
```
