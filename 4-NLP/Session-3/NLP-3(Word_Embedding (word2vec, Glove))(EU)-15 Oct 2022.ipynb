## Word_embedding
- Word embeddings are numeric vectors that show semantic relationships between words. To see how close the two words are to each other semantically, the cosinus similarity of these word vectors is checked. The closer the value is to 1, the closer the words are semantically.

## Word2Vec
- Word2Vec is a statistical method for efficiently learning a word embedding from a text corpus. It was developed by Google in 2013. There are 2 different algorithms for training in Word2vec. These are the Continous Bag-of-Words(CBOW model) and the Skip-Gram Model.

    - In the **CBOW model**, a word (middle word) in the sentence is tried to be predicted based on the words around that word. (By giving the words around it as input to the model, the middle word is tried to be predicted. The number of words around the middle word is determined by windows size parameter and usually, the numbers 5 or 10 are chosen.
    
    - In the **Skip-Gram model**, based on a word (middle word) in the sentence, the words around the middle word are tried to be predicted. (The words around the middle word are tried to be predicted by giving the middle word as an input to the model. The number of words around the middle word is determined by the **windows size** parameter and usually, the numbers 5 or 10 are chosen.)
    - **Global Vectors (Glove):** It was developed by Stanford University in 2014. It is a model that takes into account the statistics of using all the words in the corpus with other words. (in the corpus, the word red is used together with the word car 100 times, etc.)
    - **Embeddings from Language Models(ELMo):** The word2vec and glove models are unfortunately unsuccesful in capturing homophones(sesteş) with different meanings. However, the Elmo model is quite succesful in capturing the semantic differences between words.
        - Example : 
            - I love the apple that my mother bought from the market.
            - I love the apple, the biggest company in the world.
        - Word2vec and Glove modules can't capture the difference in semantics between the apple words in the two different sentences above. However, the ELMo model will capture this difference.

```python
import warnings
warnings.filterwarnings("ignore")

from gensim.models import Word2Vec
```

```python
import zipfile

text = zipfile.ZipFile('newspaper.zip')
text = text.read('newspaper.txt')
text = text.decode('utf-8')

text[:1000]
# zipli dosyayı okumak için kullanabilirisniz
```

```python
#text = open('newspaper.txt', 'r', encoding='utf8')
#text = text.read()
#text[:1000]

# dosyayı zip içerisinden çıkarsaydık. Bu formatı kullanacaktık.
```

```python
print(text[:1000])
```

```python
list_sent = text.split('\n')
list_sent[:10]
```

```python
list_sent[0].split()
```

```python
corpus = []

for sent in list_sent:
    corpus.append(sent.split())
```

```python
print(corpus[:10])
```

```python
model = Word2Vec(corpus, vector_size=100, window=5, min_count=5, sg=1)
```

```python
model.wv['ankara']
```

```python
model.wv.most_similar('öğretmen')
```

```python
model.wv.most_similar('kırmızı')
```

```python
model.wv.most_similar('eve')
```

```python
model.wv.most_similar('mavi')
```

```python
model.wv.most_similar(positive=['öğrenme', 'doktor'], negative=['tedavi'], topn=5)
```

```python
model.wv.most_similar(positive=['ankara', 'belçika'], negative=['brüksel'], topn=1)
```

```python
model.save("word2vec.model")
```
