## Cleaning Data

```python
import re
s = "http\\:www.mynet.com #lateflight @airlines"
s = re.sub("http\S+", "", s).strip() # S+ >> Boşluk hariç tüm karakterler en az bir kere olacak anlamındadır.
s                                    # Baştaki ve sondaki boşlukları almak için de strip() kullandık.
                                     # S* >> boşluk hariç tüm karakterler 0 kez veya daha fazla olacak anlamındadır.
```

```python
s = re.sub("#\S+", "", s)
s
```

```python
s = re.sub("@\S+", "", s)
s.strip()
```

```python
stop_words = stopwords.words('english')

for i in ["not", "no"]:
        stop_words.remove(i)
# duygu analizi yapacağımız için stop_words' lerden not ve no'yu çıkarttık.
```

```python
def cleaning(data):
    
    import re
    
    #1. Removing URLS
    data = re.sub('http\S+', '', data).strip()
    data = re.sub('www\S+', '', data).strip()

    #2. Removing Tags
    data = re.sub('#\S+', '', data).strip()

    #3. Removing Mentions
    data = re.sub('@\S+', '', data).strip()
    
    #4. Removing upper brackets to keep negative auxiliary verbs in text
    data = data.replace("'", "")
    
    #5. Tokenize
    text_tokens = word_tokenize(data.lower())
    
    #6. Remove Puncs and number
    tokens_without_punc = [w for w in text_tokens if w.isalpha()]
    
    #7. Removing Stopwords
    tokens_without_sw = [t for t in tokens_without_punc if t not in stop_words]
    
    #8. lemma
    text_cleaned = [WordNetLemmatizer().lemmatize(t) for t in tokens_without_sw]
    
    #joining
    return " ".join(text_cleaned)
```

```python
cleaning_text = df2["text"].apply(cleaning)
cleaning_text.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df2["text"]
y= df2["airline_sentiment"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=101)
# Skore iyi çıkmazsa ilk akla gelecek şey test datasında ignore edilen datalar mı var? olmalı..
# Ve test_size 'ı küçültüp daha çok data ile train yapmayı sağlayın...
```

## Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(preprocessor=cleaning, min_df=3) # ngram_range=(1,2), max_features= 1500
X_train_count = vectorizer.fit_transform(X_train)
X_test_count = vectorizer.transform(X_test)

# min_df = 3    --> corpusumda 3'den daha az document'da (yorumda) geçen tokenleri ignore et.(çoğunlukla 3-4-5 sayıları kullnlr)
# max_df = 1000 --> corpusumda 1000'den daha fazla document'da (yorumda) geçen tokenleri ignore et. (0.95, 1.0)
    # üst sınırları belirlemek her zaman riskli ve tehlikelidir.
    # max_df ve max_features' u kullanmanız önerilmez! 
# max_features = 1500 --> corpusumda en sık kullanılan ilk 1500 feature (tokeni) dikkate al diğerlerini ignore et.
# ngram_range = (1,2) olursa tokenleri hem birer bier hem de ikişer ikişer kelime grupları halinde alır. 
    # tamlamaları yakalamak için kullanılır.
    # best practice olarak tavisye edilen (1,2)'de kalmak daha ileri geçmemektir.
    # olumlu ve olumsuz kalıplar çoğunlukla 2 kelimeli isim ve sıfat tamlamalarından oluşur.
    
```

## Model Comparisons - Vectorization

```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score, recall_score
```

```python
def eval(model, X_train, X_test):
    y_pred = model.predict(X_test)
    y_pred_train = model.predict(X_train)
    print(confusion_matrix(y_test, y_pred))
    print("Test_Set")
    print(classification_report(y_test,y_pred))
    print("Train_Set")
    print(classification_report(y_train,y_pred_train))
    
```

```python
# Naive Bayes genelde NLP'de iyi sonuç verdiği için  NLP'de kullanılan bir ML algoritmasıdır.
# A olayının gerçekleşmesi durumunda B olayının gerçeleşmesi durumu'nun hesaplanması bu algoritmaya örnektir.
# Aşağıda daha detaylı bir örnekte anlatılıyor.
```

## Naive Bayes

https://medium.com/analytics-vidhya/na%C3%AFve-bayes-algorithm-5bf31e9032a2

```python
from sklearn.naive_bayes import MultinomialNB, BernoulliNB # BernoulliNB for binary model
# normalde multiclass'da multinomial, binary'de bernoulli kullanılması önerilir ancak ikisini de kullanın.
    #hangisi yüksek skor veriyorsa onu tercih edin.
```

```python
nb = MultinomialNB(alpha=10)
nb.fit(X_train_count,y_train)
# Harika TV kötü kargo deseydik harika kelimesi yorumlarda hiç geçmediği için tüm olasılığı 0' a çekecekti.
#Bunu önelemek için alpha değeri kullanıyoruz.
# paya alpha değeri eklerken paydaya 10 * olumlu yorum sayısı ekler. Böylece olasılığı 0'dan kurtarır.
# Overfitting problemini önlemek için alpha'yı kullanırız.
# Bu alpha değeri tüm tokenlere eklenir...
```

```python
print("NB MODEL")
eval(nb, X_train_count, X_test_count)
```

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score

scoring = {'accuracy': make_scorer(accuracy_score),
            'precision-neg': make_scorer(precision_score, average=None, labels=["negative"]),
            'recall-neg': make_scorer(recall_score, average=None, labels = ["negative"]),
            'f1-neg': make_scorer(f1_score, average=None, labels = ["negative"])}

model = MultinomialNB(alpha=10)
scores = cross_validate(model, X_train_count, y_train, scoring = scoring, cv = 10, return_train_score=True)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]
```

```python
from yellowbrick.classifier import PrecisionRecallCurve
# Data imbalance olduğu için PrecisionRecallCurve' e bakacağız.

viz = PrecisionRecallCurve(
    MultinomialNB(alpha=10),
    classes=nb.classes_,
    per_class=True, # Her sınıf için yazmasını istiyoruz.
    cmap="Set1"
)
viz.fit(X_train_count,y_train)
viz.score(X_test_count, y_test)
viz.show();

# model negatif labelları diğer labellardan %92 başarı oranı ile ayırabiliyor.
```

```python
y_pred = nb.predict(X_test_count)
nb_count_rec_neg = recall_score(y_test, y_pred, labels = ["negative"], average = None)
nb_count_f1_neg = f1_score(y_test, y_pred, labels = ["negative"], average = None)
nb_AP_neg = viz.score_["negative"]
# Sadece viz.score yapsaydık bütün skore'ları döndürecekti.
```
